```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
 
# Build a dataframe with 4 connections
df = pd.DataFrame({ 'from':['A','A', 'B', 'C','A'], 'to':['A','D', 'A', 'E','C',]})
 
# Build your graph
G=nx.from_pandas_edgelist(df, 'from', 'to')
 
# Plot it
nx.draw(G, with_labels=True)
plt.show()
```

```python
import os
import json
ignore_files = ['activity_tree.py', 'flow_graph.ipynb','.DS_Store']

files = os.listdir()
files = [x for x in files if x not in ignore_files]

count = 1
for file in files:
    try:
        from_lst = []
        to_lst = []
        with open(file,'r') as f:
            print(file)
            activity_matching_dict = json.load(f)
            for k,vs in activity_matching_dict.items():
                if vs == [""]:
                    continue
                for v in vs:
                    from_lst.append(k.split('/')[-1])
                    to_lst.append(v.split('.')[-1])
                count += 1
#                 if count>10:
#                     break
#                 break   
                    
            df = pd.DataFrame({ 'from':from_lst, 'to':to_lst})

            # Build your graph
            G=nx.from_pandas_edgelist(df, 'from', 'to')

            # Plot it
            nx.draw(G, with_labels=True)
            # Plot the graph without labels
#             nx.draw(G)
            plt.show()
#         break
    except:
        print(file)
```
